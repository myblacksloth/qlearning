## algoritmi di ordinamento

```python
def minimo(l): # << lista O(n)
    from math import inf
    min = inf
    for i in range(len(l)):
        if l[i] < min:
            min = l[i]
    return min
# #O(n)
```

```python
# O(n^2)
def ordina(l): # << lista
    for i in range(len(l)):
        for j in range(len(l)):
            if l[i] < l[j]:
                l[i] , l[j] = l[j], l[i]
    return l
```

```python
def heapsort(v): # << array O(n logn)

    def heapify(v, n , i): # << array, n=dimensione, i=indice della radice dell'heap
        largest = i # all'inizio supponiamo che il nodo piu' grande sia la radice
        l = 2 * i + 1 # indice del figlio sinitro di i
        r = 2 * i + 2 # indice del figlio destro di i

        # se il figlio sinistro esiste ed e' piu' grande del nodo [radice]
        if l<n and v[i]<v[l]:
            largest = l
        
        # se il figlio sinistro esiste ed e' piu' grande del nodo piu' grande [radice o figlio sx]
        if r<n and v[largest]<v[r]:
            largest = r
        
        # modifica la radice se necessario
        if largest != i:
            v[i], v[largest] = v[largest], v[i]

            # effettua nuovamente heapify per la nuova radice
            heapify(v, n, largest)

    n = len(v) # dimensione dell'heap

    for i in range(n//2 - 1, -1, -1):
        heapify(v, n, i)
    
    for i in range(n-1, 0, -1):
        v[i], v[0] = v[0], v[i]
        heapify(v, i, 0)
```

```python
def insertionSort(l): # << array O(n logn)
    for i in range(1, len(l)):
        k = l[i]
        j = i-1
        while j>=0 and k<l[j]:
            l[j+1] = l[j]
            j-=1
        l[j+1] = k
```

```python
def mergeSort(l):
    if len(l)>1:
        mid = len(l)//2
        L = l[:mid]
        R = l[mid:]
        mergeSort(L)
        mergeSort(R)
        i = j = k = 0
        while i<len(L) and j < len(R):
            if L[i] < R[j]:
                l[k] = L[i]
                i+=1
            else:
                l[k] = R[j]
                j+=1
            k+=1
        while i<len(L):
            l[k] = L[i]
            i+=1
            k+=1
        while j < len(R):
            l[k] = R[j]
            j+=1
            k+=1
```

## exe 1

caso random

```python
durata1 = []
for k in progressbar(range(3, 250)):
    durata = 0
    for _ in range(100):
        l = []
        for i in range(k):
            l.append(random.randint(1, 999))
        # 
        start = time()
        # 
        minimo(l)
        # 
        durata = durata + (time() - start)
    # salvo la media delle durate
    durata1.append(durata/100)
```

caso peggiore

```python
durata2 = []
for k in progressbar(range(3, 250)):
    durata = 0
    for _ in range(100):
        l = []
        for i in range(k):
            l.append(random.randint(1, 999))
        l.sort(reverse=True)
        # 
        start = time()
        # 
        minimo(l)
        # 
        durata = durata + (time() - start)
    # salvo la media delle durate
    durata2.append(durata/100)
```

cso migliore

```python
durata3 = []
for k in progressbar(range(3, 250)):
    durata = 0
    for _ in range(100):
        l = []
        for i in range(k):
            l.append(random.randint(1, 999))
        l.sort()
        # 
        start = time()
        # 
        minimo(l)
        # 
        durata = durata + (time() - start)
    # salvo la media delle durate
    durata3.append(durata/100)
```

risultati

```python
plt.figure(figsize=(20,10))
plt.plot(durata1, color='b')
plt.plot(durata2, color='r')
plt.plot(durata3, color='g')
plt.grid() # mostra la griglia
```
